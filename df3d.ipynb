```python
%load_ext autoreload
%autoreload 2

from pyba.CameraNetwork import CameraNetwork
import pickle
import glob
import numpy as np
# TODO need to set joints to zero if invisible

n_cameras = 7
image_path = './data/test/camera_{cam_id}_img_00000{img_id}.jpg'
pr_path = './data/test/df3d_2/pose_result*.pkl'
d = pickle.load(open(glob.glob(pr_path)[0], 'rb'))
points2d = d['points2d']
calib = {cid:d[cid] for cid in range(n_cameras)}
bones = np.array([[i, i+1] for i in range(15) if (i+1)%5!=0])
bones = np.concatenate([bones, bones + 19])
camNet = CameraNetwork(points2d=points2d, calib=calib, image_path=image_path, bones=bones)
```

```python
_ = camNet.triangulate()
```

```python
camNet.reprojection_error()
```

```python
import matplotlib.pyplot as plt
img = camNet.plot_2d(0, points='points2d')
plt.figure(figsize=(20,20))
plt.imshow(img, cmap='gray')
plt.axis('off')
```
